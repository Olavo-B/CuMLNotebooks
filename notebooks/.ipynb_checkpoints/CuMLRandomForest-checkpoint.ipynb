## Rodar Random Forest

```python
def run_random_forest(new_data, target, _n_estimators, _max_depth, debug=False):
      from cuml.ensemble import RandomForestClassifier
      from cuml.metrics import accuracy_score
      from cuml.model_selection import train_test_split
      import cudf
      import time
      import gc


      # Converter os dados para um DataFrame do cudf
      X_cudf = new_data
      y_cudf = cudf.Series(target)

      # Dividir os dados em conjuntos de treinamento e teste
      X_train, X_test, y_train, y_test = train_test_split(X_cudf, y_cudf, test_size=0.2, random_state=42)


      # Instanciar e treinar o modelo RandomForest
      rf = RandomForestClassifier(n_estimators=_n_estimators,
                                  max_depth = _max_depth,
                                  random_state=42)

      start_time = time.time()
      rf.fit(X_train, y_train)
      end_time = time.time()

      if debug:
        print(f'Tempo de treinamento do modelo RandomForest {_n_estimators},{_max_depth}: {end_time - start_time} segundos')

      # Fazer previsões no conjunto de teste
      y_pred = rf.predict(X_test)

      del X_train, X_test, y_train, X_cudf, y_cudf
      gc.collect()

      # Calcular a acurácia
      accuracy = accuracy_score(y_test, y_pred)
      if debug:
        print(f'Acurácia do modelo RandomForest {_n_estimators},{_max_depth}: {accuracy}')

      # Limpar a memória da GPU
      del y_test, y_pred
      gc.collect()

      return accuracy, rf
```

```python
import cudf

data = cudf.DataFrame(data)

# Definindo os valores para N-trees e depth
n_trees = np.arange(5, 105, 5)  # De 5 até 100, múltiplos de 5
depth = np.arange(5, 105, 5)    # De 5 até 100, múltiplos de 5

accs = []
for n in n_trees:
    for d in depth:
        acc, rf = run_random_forest(data, target, n, d, debug=True)
        accs.append((n, d, acc))
```

```python


def data_aumentation(data: cudf.DataFrame, target: cudf.Series, expect_size: int = 10e4, debug=False):
    """
    Concatena n cópias de data e target, aplica um shuffle e retorna um DataFrame contendo
    tanto os dados quanto os alvos, garantindo que estejam alinhados.

    Parameters:
    data (cudf.DataFrame): O DataFrame de entrada contendo as características.
    target (cudf.Series): O Series contendo o alvo correspondente aos dados.
    expect_size (int, optional): O tamanho esperado para o conjunto de dados aumentado. O padrão é 100.000.
    debug (bool, optional): Exibe mensagens de debug se True. O padrão é False.

    Returns:
    cudf.DataFrame: DataFrame contendo os dados aumentados e os alvos alinhados.
    """

    import cudf
    import math

    #Verifica se data e target são cudf
    if not isinstance(data, cudf.DataFrame):
        raise ValueError('data must be a cudf.DataFrame')
    target = cudf.Series(target) if not isinstance(target, cudf.Series) else target

    # Verifica o número de cópias necessárias para atingir o tamanho esperado
    num_copies = math.ceil(expect_size / len(data))

    # Concatena as cópias de data e target
    augmented_data = cudf.concat([data] * num_copies, ignore_index=True)
    augmented_target = cudf.concat([target] * num_copies, ignore_index=True)

    # Combina data e target em um único DataFrame
    augmented_data['target'] = augmented_target

    # Embaralha o DataFrame combinado
    augmented_data = augmented_data.sample(frac=1).reset_index(drop=True)

    # Trunca para o tamanho exato esperado, caso seja maior
    if len(augmented_data) > expect_size:
        augmented_data = augmented_data.iloc[:expect_size]

    augmentad_target = augmented_data['target']
    augmented_data = augmented_data.drop('target', axis=1)

    if debug:
        print(f'Augmented data size: {len(augmented_data)}')




    return augmented_data, augmentad_target
```

```python
from time import time
from cuml.metrics import accuracy_score

data_aume, target_aumentation = data_aumentation(data, target, 10000000, True) # EDITAR PARA MODIFICAR O TAMANHO DO DATASET



time_start = time()
y_pred  = rf.predict(data_aume)
time_end = time()




accuracy = accuracy_score(target_aumentation, y_pred)
print(f'Acurácia do modelo RandomForest: {accuracy}')
print(f'Tempo de classificação do modelo RandomForest: {time_end - time_start} segundos')
```
